```python
import numpy as np
from scipy.sparse import diags, csr_matrix, lil_matrix
from scipy.optimize import root
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from dataclasses import dataclass
from typing import Tuple, Optional, Union
from scipy.sparse.linalg import spsolve

# Constants
R = 8.31446261815324        # universal gas constant [J/(mol·K)]
FARADAY = 96485.33212       # Faraday constant [C/mol]
F = FARADAY                 # alias for convenience


def mg_per_cm2_to_kg_per_m2(x: float) -> float:
    """Convert mass loading from mg/cm² to kg/m²."""
    return float(x) * 1e-2  # 1 mg/cm² = 0.01 kg/m²


@dataclass
class Parameters:
    """Container for all model parameters."""
    # Temperature and pressure
    T: float              # cell temperature [K]
    P_anode: float        # absolute pressure at anode gas channel [Pa]
    P_cathode: float      # absolute pressure at cathode gas channel [Pa]

    # Layer thicknesses [m]
    L_PTL: float
    L_aCL: float
    L_mem: float
    L_cCL: float
    L_GDL: float

    # Catalyst layer loading and mass fractions
    loading_aCL: float    # [mg/cm²]
    loading_cCL: float    # [mg/cm²]
    w_cat_aCL: float      # fraction of anode catalyst mass
    w_ion_aCL: float      # fraction of anode ionomer mass
    w_cat_cCL: float      # fraction of cathode catalyst mass
    w_ion_cCL: float      # fraction of cathode ionomer mass

    # Specific surface areas [m²/g]
    SA_IrOx: float
    SA_PtC: float

    # Densities [kg/m³]
    ρ_IrO2: float
    ρ_Nafion: float
    ρ_PtC: float

    # Kinetic parameters at reference temperature (T_ref = 353 K)
    j0_aCL_ref: float     # exchange current density of anode [A/m²]
    j0_cCL_ref: float     # exchange current density of cathode [A/m²]
    b_aCL_ref: float      # Tafel slope of anode [V/decade]
    b_cCL_ref: float      # Tafel slope of cathode [V/decade]
    E_a_j0_aCL: float     # activation energy of anode exchange current [J/mol]
    E_a_j0_cCL: float     # activation energy of cathode exchange current [J/mol]
    E_a_b_aCL: float      # activation energy of anode Tafel slope [J/mol]
    E_a_b_cCL: float      # activation energy of cathode Tafel slope [J/mol]

    # Conductivities [S/m]
    σ_e_aCL: float
    σ_e_cCL: float
    σ_e_PTL: float
    σ_e_GDL: float
    σ_m_ref: float        # protonic conductivity at reference temperature [S/m]

    # Mass transport coefficients [s⁻¹]
    k_H2_mass: float
    k_O2_mass: float

    # Diffusion coefficients in liquid water at reference temperature [m²/s]
    D_H2_ref: float
    D_O2_ref: float
    E_a_D_H2: float       # activation energy for H₂ diffusion [J/mol]
    E_a_D_O2: float       # activation energy for O₂ diffusion [J/mol]

    # Solubilities of gases in water [mol/(m³·Pa)]
    S_H2_ref: float       # reference hydrogen solubility
    S_O2_ref: float       # reference oxygen solubility
    S_H2: float           # solubility of hydrogen at operating temperature
    S_O2: float           # solubility of oxygen at operating temperature

    # Electro‑osmotic drag coefficient
    n_drag: float

    # Hydraulic permeabilities [m²]
    K_PTL: float
    K_aCL: float
    K_cCL: float
    K_GDL: float

    # Contact angles [°]
    θ_PTL: float
    θ_aCL: float
    θ_cCL: float
    θ_GDL: float

    # Membrane water content and equivalent weight
    λ: float
    EW: float
    a_interface_aCL: float
    a_interface_cCL: float
    w_C_in_PtC: float = 0.5


def S_H2_pure_water(T: float) -> float:
    """
    H2 solubility in water [mol/(m³·Pa)] from SI equation S.15.
    Valid from 273-353 K. Pre-factor 0.54 converts from mole fraction.
    """
    return 0.54 * np.exp(-48.1611 + 5528.45/T + 16.8893*np.log(T/100.0))

def S_O2_pure_water(T: float) -> float:
    """
    O2 solubility in water [mol/(m³·Pa)] from SI equation S.16.
    Valid from 273-353 K. Pre-factor 0.54 converts from mole fraction.
    """
    return 0.54 * np.exp(-66.7354 + 8477.71/T + 24.4526*np.log(T/100.0))

def D_H2_water(T: float) -> float:
    """Diffusion coefficient of dissolved hydrogen in water [m²/s]."""
    return 4.9e-6 * np.exp(-16560.885/(R*T))


def D_O2_water(T: float) -> float:
    """Diffusion coefficient of dissolved oxygen in water [m²/s]."""
    return 4.2e-6 * np.exp(-18367.527/(R*T))

def protonic_conductivity_at_T(λ: float, T: float, T_ref: float = 353.0) -> float:
    """
    Nafion protonic conductivity at temperature T using Springer correlation.
    SI equation S.11: σ_P = (0.5139*λ - 0.326) * exp(1268*(1/303 - 1/T))
    """
    σ_P_ref = (0.5139 * λ - 0.326) * np.exp(1268.0 * (1.0/303.0 - 1.0/T_ref))
    return σ_P_ref * np.exp(1268.0 * (1.0/T_ref - 1.0/T))

    
def default_parameters(T: float = 353.0,
                       membrane: str = 'N117',
                       pressure: float = 1e5) -> Parameters:
    """Return default model parameters from Franz et al."""
    # Layer thicknesses [m]
    L_PTL = 250e-6
    L_aCL = 7e-6
    L_cCL = 6e-6
    L_GDL = 250e-6

    # Membrane swelling calculation (SI equation S.5, S.10)
    λ_ion = 21.0
    M_H2O = 0.018  # kg/mol
    ρ_water = 970.0  # kg/m³ at 80°C
    EW = 1.1  # kg/mol
    ρ_Nafion_dry = 2000.0  # kg/m³

    V_ionomer_dry_mol = EW / ρ_Nafion_dry
    V_water_mol = M_H2O / ρ_water
    swelling_factor = 1.0 + (λ_ion * V_water_mol) / V_ionomer_dry_mol
    
    if membrane == 'N212':
        L_mem_dry = 45e-6
    else:  # N117
        L_mem_dry = 178e-6
    
    L_mem = L_mem_dry * swelling_factor

    # Catalyst loadings [mg/cm²]
    loading_aCL = 1.0
    loading_cCL = 0.2

    # Mass fractions
    w_cat_aCL = 0.7
    w_ion_aCL = 0.3
    w_cat_cCL = 0.85
    w_ion_cCL = 0.15

    # Specific surface areas [m²/g] - Table S1
    SA_IrOx = 30.0
    SA_PtC = 120.0

    # Densities [kg/m³]
    ρ_IrO2 = 11.66e3
    ρ_Nafion = ρ_Nafion_dry  # Use same value
    ρ_PtC = 3.0e3

    # Kinetic parameters - Table S1, reference T = 353 K
    j0_aCL_ref = 1.7e-8   # A/m²
    j0_cCL_ref = 2.5e3    # A/m²
    b_aCL_ref  = 44e-3    # V/dec (0.044 V/dec)
    b_cCL_ref  = 135e-3   # V/dec (0.135 V/dec)

    # Activation energies [J/mol] - Table S1
    E_a_j0_aCL = 0.6 * F  # 57891 J/mol
    E_a_j0_cCL = 0.3 * F  # 28946 J/mol
    E_a_b_aCL  = 0.0
    E_a_b_cCL  = 0.0

    # Electronic conductivities [S/m] - Table S1
    σ_e_PTL = 490.0
    σ_e_aCL = 490.0
    σ_e_cCL = 450.0
    σ_e_GDL = 450.0

    # Proton conductivity at 353 K (Springer correlation, SI eq S.11)
    σ_m_ref = (0.5139 * λ_ion - 0.326) * np.exp(1268.0 * (1.0/303.0 - 1.0/353.0))

    # Mass transfer coefficients [s⁻¹] - fitted to experiments
    k_H2_mass = 540.0   # Low value for high crossover case
    k_O2_mass = 3500.0  # Assumed similar magnitude

    # Diffusion activation energies [J/mol] - SI eq S.13, S.14
    E_a_D_H2 = 16560.885
    E_a_D_O2 = 18367.527
    
    # Compute reference diffusion coefficients at 353 K
    T_ref = 353.0
    D_H2_ref = 4.9e-6 * np.exp(-E_a_D_H2 / (R * T_ref))
    D_O2_ref = 4.2e-6 * np.exp(-E_a_D_O2 / (R * T_ref))

    # Solubilities (use functions for temperature dependence)
    S_H2_ref = S_H2_pure_water(T_ref)
    S_O2_ref = S_O2_pure_water(T_ref)
    S_H2_val = S_H2_pure_water(T)
    S_O2_val = S_O2_pure_water(T)

    # Electro-osmotic drag - Table S1
    n_drag = 2.5

    # Hydraulic permeabilities [m²] - Table S1
    K_PTL = 6.4e-14
    K_aCL = 5.0e-17
    K_cCL = 5.0e-17
    K_GDL = 2.0e-15

    # Contact angles [°] - Table S1
    θ_PTL = 30.0
    θ_aCL = 80.0
    θ_cCL = 80.0
    θ_GDL = 80.0

    # Interface area density [m²/m³] - typical order of magnitude
    a_interface_aCL = 1.0e6
    a_interface_cCL = 1.0e6
    
    # Carbon mass fraction in Pt/C catalyst
    w_C_in_PtC = 0.5

    return Parameters(
        T=T, P_anode=pressure, P_cathode=pressure,
        L_PTL=L_PTL, L_aCL=L_aCL, L_mem=L_mem, L_cCL=L_cCL, L_GDL=L_GDL,
        loading_aCL=loading_aCL, loading_cCL=loading_cCL,
        w_cat_aCL=w_cat_aCL, w_ion_aCL=w_ion_aCL, w_cat_cCL=w_cat_cCL, w_ion_cCL=w_ion_cCL,
        SA_IrOx=SA_IrOx, SA_PtC=SA_PtC,
        ρ_IrO2=ρ_IrO2, ρ_Nafion=ρ_Nafion, ρ_PtC=ρ_PtC,
        j0_aCL_ref=j0_aCL_ref, j0_cCL_ref=j0_cCL_ref,
        b_aCL_ref=b_aCL_ref, b_cCL_ref=b_cCL_ref,
        E_a_j0_aCL=E_a_j0_aCL, E_a_j0_cCL=E_a_j0_cCL,
        E_a_b_aCL=E_a_b_aCL, E_a_b_cCL=E_a_b_cCL,
        σ_e_aCL=σ_e_aCL, σ_e_cCL=σ_e_cCL, σ_e_PTL=σ_e_PTL, σ_e_GDL=σ_e_GDL,
        σ_m_ref=σ_m_ref,
        k_H2_mass=k_H2_mass, k_O2_mass=k_O2_mass,
        D_H2_ref=D_H2_ref, D_O2_ref=D_O2_ref,
        E_a_D_H2=E_a_D_H2, E_a_D_O2=E_a_D_O2,
        S_H2_ref=S_H2_ref, S_O2_ref=S_O2_ref, S_H2=S_H2_val, S_O2=S_O2_val,
        n_drag=n_drag,
        K_PTL=K_PTL, K_aCL=K_aCL, K_cCL=K_cCL, K_GDL=K_GDL,
        θ_PTL=θ_PTL, θ_aCL=θ_aCL, θ_cCL=θ_cCL, θ_GDL=θ_GDL,
        λ=λ_ion, EW=EW, w_C_in_PtC=w_C_in_PtC,
        a_interface_aCL=a_interface_aCL, a_interface_cCL=a_interface_cCL
    )


@dataclass
class Mesh:
    """Discretisation of the through‑plane direction."""
    z: np.ndarray
    Δz: np.ndarray
    n: int
    idx_PTL: range
    idx_aCL: range
    idx_mem: range
    idx_cCL: range
    idx_GDL: range


def create_mesh(params: Parameters, n_PTL: int = 100, n_aCL: int = 100,
                n_mem: int = 100, n_cCL: int = 100, n_GDL: int = 100) -> Mesh:
    """Create a uniform mesh for finite–volume discretisation."""
    Δz_PTL = params.L_PTL / n_PTL
    Δz_aCL = params.L_aCL / n_aCL
    Δz_mem = params.L_mem / n_mem
    Δz_cCL = params.L_cCL / n_cCL
    Δz_GDL = params.L_GDL / n_GDL

    z_PTL = np.array([Δz_PTL/2 + i*Δz_PTL for i in range(n_PTL)])
    z_aCL = np.array([params.L_PTL + Δz_aCL/2 + i*Δz_aCL for i in range(n_aCL)])
    z_mem = np.array([params.L_PTL + params.L_aCL + Δz_mem/2 + i*Δz_mem for i in range(n_mem)])
    z_cCL = np.array([params.L_PTL + params.L_aCL + params.L_mem + Δz_cCL/2 + i*Δz_cCL for i in range(n_cCL)])
    z_GDL = np.array([params.L_PTL + params.L_aCL + params.L_mem + params.L_cCL + Δz_GDL/2 + i*Δz_GDL for i in range(n_GDL)])

    z = np.concatenate([z_PTL, z_aCL, z_mem, z_cCL, z_GDL])
    Δz = np.concatenate([np.full(n_PTL, Δz_PTL), np.full(n_aCL, Δz_aCL), np.full(n_mem, Δz_mem),
                        np.full(n_cCL, Δz_cCL), np.full(n_GDL, Δz_GDL)])
    n = len(z)

    idx_PTL = range(0, n_PTL)
    idx_aCL = range(n_PTL, n_PTL+n_aCL)
    idx_mem = range(n_PTL+n_aCL, n_PTL+n_aCL+n_mem)
    idx_cCL = range(n_PTL+n_aCL+n_mem, n_PTL+n_aCL+n_mem+n_cCL)
    idx_GDL = range(n_PTL+n_aCL+n_mem+n_cCL, n)

    return Mesh(z, Δz, n, idx_PTL, idx_aCL, idx_mem, idx_cCL, idx_GDL)


@dataclass
class Solution:
    """Container for the steady–state solution fields."""
    φ_s: np.ndarray
    φ_m: np.ndarray
    c_H2: np.ndarray
    c_O2: np.ndarray
    I: np.ndarray
    converged: bool
    iterations: int


def temperature_correction(value_ref: float, E_a: float,
                          T_ref: float, T: float) -> float:
    """Apply an Arrhenius temperature correction."""
    if E_a == 0.0:
        return value_ref
    return value_ref * np.exp(-E_a / R * (1.0/T - 1.0/T_ref))


def compute_volume_fractions(params: Parameters, mesh: Mesh):
    """Compute volume fractions for each layer."""
    n = mesh.n
    ε_cat = np.zeros(n)
    ε_ion = np.zeros(n)
    ε_pore = np.zeros(n)
    ε_H2O_ion = np.zeros(n)

    # SI-consistent constants
    M_H2O = 0.018      # kg/mol
    ρ_water = 970.0    # kg/m³
    ρ_ion_dry = params.ρ_Nafion

    EW = params.EW if params.EW < 10 else params.EW * 1e-3

    V_ionomer_dry_mol = EW / ρ_ion_dry
    V_water_mol = M_H2O / ρ_water

    swelling = 1.0 + (params.λ * V_water_mol) / V_ionomer_dry_mol
    εH2O_single = (params.λ * V_water_mol) / (params.λ * V_water_mol + V_ionomer_dry_mol)

    # aCL
    L_aCL = mg_per_cm2_to_kg_per_m2(params.loading_aCL)
    for i in mesh.idx_aCL:
        V_cat = (L_aCL * params.w_cat_aCL) / params.ρ_IrO2
        V_ion = (L_aCL * params.w_ion_aCL) / ρ_ion_dry * swelling
        ε_cat[i] = V_cat / params.L_aCL
        ε_ion[i] = V_ion / params.L_aCL
        ε_pore[i] = 1.0 - ε_cat[i] - ε_ion[i]
        ε_H2O_ion[i] = εH2O_single

    # cCL
    ρ_Pt = 21400.0
    ρ_C = 1800.0
    w_C_in_PtC = params.w_C_in_PtC
    ρ_PtC = 1.0 / (w_C_in_PtC/ρ_C + (1.0 - w_C_in_PtC)/ρ_Pt)

    L_cCL = mg_per_cm2_to_kg_per_m2(params.loading_cCL)
    for i in mesh.idx_cCL:
        V_cat = (L_cCL * params.w_cat_cCL) / ρ_PtC
        V_ion = (L_cCL * params.w_ion_cCL) / ρ_ion_dry * swelling
        ε_cat[i] = V_cat / params.L_cCL
        ε_ion[i] = V_ion / params.L_cCL
        ε_pore[i] = 1.0 - ε_cat[i] - ε_ion[i]
        ε_H2O_ion[i] = εH2O_single

    # PTL, GDL porosities
    ε_pore[mesh.idx_PTL] = 0.75
    ε_pore[mesh.idx_GDL] = 0.60

    # membrane
    ε_cat[mesh.idx_mem] = 0.0
    ε_ion[mesh.idx_mem] = 1.0
    ε_pore[mesh.idx_mem] = 0.0
    ε_H2O_ion[mesh.idx_mem] = εH2O_single

    return ε_cat, ε_ion, ε_pore, ε_H2O_ion


def effective_conductivities(params: Parameters, mesh: Mesh) -> Tuple[np.ndarray, np.ndarray]:
    """Compute effective electronic and protonic conductivities."""
    ε_cat, ε_ion, ε_pore, _ = compute_volume_fractions(params, mesh)
    n = mesh.n
    σ_s = np.zeros(n)
    σ_m = np.zeros(n)

    # Protonic conductivity at operating temperature
    σ_m_Tref = params.σ_m_ref
    σ_m_cell = protonic_conductivity_at_T(params.λ, params.T)

    for i in range(n):
        if i in mesh.idx_PTL:
            σ_s[i] = params.σ_e_PTL
            σ_m[i] = 1e-12
        elif i in mesh.idx_aCL:
            σ_s[i] = params.σ_e_aCL * (ε_cat[i])**1.5
            σ_m[i] = σ_m_cell * (ε_ion[i])**1.5
        elif i in mesh.idx_mem:
            σ_s[i] = 1e-12
            σ_m[i] = σ_m_cell
        elif i in mesh.idx_cCL:
            σ_s[i] = params.σ_e_cCL * (ε_cat[i])**1.5
            σ_m[i] = σ_m_cell * (ε_ion[i])**1.5
        else:  # GDL
            σ_s[i] = params.σ_e_GDL
            σ_m[i] = 1e-12

    return σ_s, σ_m


def effective_diffusion(params: Parameters, mesh: Mesh) -> Tuple[np.ndarray, np.ndarray]:
    """Effective diffusion coefficients of dissolved H2 and O2."""
    n = mesh.n
    D_H2 = np.zeros(n)
    D_O2 = np.zeros(n)

    # Temperature-dependent molecular diffusion
    T_ref = 353.0
    D_H2_base = temperature_correction(params.D_H2_ref, params.E_a_D_H2, T_ref, params.T)
    D_O2_base = temperature_correction(params.D_O2_ref, params.E_a_D_O2, T_ref, params.T)

    ε_cat, ε_ion, ε_pore, ε_H2O_ion = compute_volume_fractions(params, mesh)

    for i in range(n):
        if i in mesh.idx_PTL or i in mesh.idx_GDL:
            # No dissolved-gas transport in PTL/GDL
            D_H2[i] = 0.0
            D_O2[i] = 0.0
        elif i in mesh.idx_aCL or i in mesh.idx_cCL:
            # Deff = D * ε_H2O,ion * ε_ion^(1.5)
            De_mult = float(ε_H2O_ion[i]) * float(ε_ion[i])**1.5
            D_H2[i] = D_H2_base * De_mult
            D_O2[i] = D_O2_base * De_mult
        else:  # Membrane
            De_mult = float(ε_H2O_ion[i])
            D_H2[i] = D_H2_base * De_mult
            D_O2[i] = D_O2_base * De_mult

    return D_H2, D_O2


def build_conduction_matrix(mesh: Mesh, σ: np.ndarray) -> csr_matrix:
    """Build finite-volume conduction matrix."""
    n = mesh.n
    row, col, val = [], [], []
    
    for i in range(1, n-1):
        # East face transmissibility
        if σ[i] > 0 and σ[i+1] > 0:
            T_e = 1.0 / ((mesh.Δz[i]*0.5)/σ[i] + (mesh.Δz[i+1]*0.5)/σ[i+1])
        else:
            T_e = 0.0
            
        # West face transmissibility  
        if σ[i] > 0 and σ[i-1] > 0:
            T_w = 1.0 / ((mesh.Δz[i]*0.5)/σ[i] + (mesh.Δz[i-1]*0.5)/σ[i-1])
        else:
            T_w = 0.0
        
        a_P = T_e + T_w
        
        if a_P > 0:
            row.extend([i, i, i])
            col.extend([i, i+1, i-1])
            val.extend([a_P, -T_e, -T_w])
        else:
            row.append(i)
            col.append(i)
            val.append(1.0)
    
    # Boundary nodes
    row.extend([0, n-1])
    col.extend([0, n-1])
    val.extend([1.0, 1.0])
    
    return csr_matrix((val, (row, col)), shape=(n, n))


def build_diffusion_matrix(mesh: Mesh, D: np.ndarray) -> csr_matrix:
    n = mesh.n
    row, col, val = [], [], []
    for i in range(1, n-1):
        Te = Tw = 0.0
        if D[i] > 0.0 and D[i+1] > 0.0:
            Te = 1.0 / ((mesh.Δz[i]*0.5)/D[i] + (mesh.Δz[i+1]*0.5)/D[i+1])
        if D[i] > 0.0 and D[i-1] > 0.0:
            Tw = 1.0 / ((mesh.Δz[i]*0.5)/D[i] + (mesh.Δz[i-1]*0.5)/D[i-1])
        aP = Te + Tw
        row.extend([i, i, i]); col.extend([i, i+1, i-1]); val.extend([aP, -Te, -Tw])
    return csr_matrix((val, (row, col)), shape=(n, n))



def kinetic_parameters(params: Parameters):
    """Temperature-corrected kinetics - properly scaled."""
    T_ref = 353.0
    

    j0_aCL_aCL = 1.7e-8   # A/m²
    j0_cCL_cCL = 2.5e3    # A/m²
    j0_a = temperature_correction(j0_vol_aCL, params.E_a_j0_aCL, T_ref, params.T)
    j0_c = temperature_correction(j0_vol_cCL, params.E_a_j0_cCL, T_ref, params.T)
    b_a = temperature_correction(params.b_aCL_ref, params.E_a_b_aCL, T_ref, params.T)
    b_c = temperature_correction(params.b_cCL_ref, params.E_a_b_cCL, T_ref, params.T)
    
    return j0_a, j0_c, b_a, b_c


def _limit_eta(eta: float, alpha: float) -> float:
    # Limit overpotential only while alpha<1 to keep Newton stable
    if alpha < 1.0:
        return np.clip(eta, -0.30, 0.30)  # safe band during continuation
    return eta

def current_density_inplace(I, params, mesh, φ_s, φ_m, c_H2, c_O2, alpha: float = 1.0):
    """Calculate volumetric current density - use effective parameters."""
    n = mesh.n
    I.fill(0.0)

    # Use EFFECTIVE exchange current densities (already volumetric)
    j0_a, j0_c, b_a, b_c = kinetic_parameters(params)
    cH2_0 = 101325.0 * params.S_H2
    cO2_0 = 101325.0 * params.S_O2

    # Anode OER
    for i in mesh.idx_aCL:
        E_rev = 1.478 - 8.347e-4 * params.T + (R * params.T) / (4.0 * F) * np.log(max(c_O2[i], 1e-30) / cO2_0)
        eta = _limit_eta(φ_s[i] - φ_m[i] - E_rev, alpha)
        exponent = np.clip(eta / b_a, -20.0, 20.0)
        I[i] = j0_a * (10.0 ** exponent)  # Already volumetric, no 'a' factor

    # Cathode HER  
    for i in mesh.idx_cCL:
        E_rev = -(R * params.T) / (2.0 * F) * np.log(max(c_H2[i], 1e-30) / cH2_0)
        eta = _limit_eta(φ_s[i] - φ_m[i] - E_rev, alpha)
        xi = (np.log(10.0) / b_c) * eta
        xi = np.clip(xi, -40.0, 40.0)
        I[i] = -j0_c * np.sinh(xi)  # Already volumetric, no 'a' factor

def current_density(params, mesh, φ_s, φ_m, c_H2, c_O2, alpha: float = 1.0) -> np.ndarray:
    I = np.zeros(mesh.n)
    current_density_inplace(I, params, mesh, φ_s, φ_m, c_H2, c_O2, alpha=alpha)
    return I


def check_physical_solution(sol: Solution, mesh: Mesh, verbose: bool = False) -> bool:
    arrays_ok = all(np.all(np.isfinite(a)) for a in (sol.φ_s, sol.φ_m, sol.c_H2, sol.c_O2, sol.I))
    if not arrays_ok:
        if verbose: print("  Non-finite values in solution arrays.")
        return False

    if np.any(sol.c_H2 < 0) or np.any(sol.c_O2 < 0):
        if verbose: print("  Negative concentrations detected.")
        return False

    j_total = np.sum(sol.I * mesh.Δz) / 1e4
    if not np.isfinite(j_total) or abs(j_total) > 20:
        if verbose: print(f"  Unphysical total current: {j_total}")
        return False

    if np.any(sol.φ_s > 5.0) or np.any(sol.φ_s < -1.0):
        if verbose: print(f"  Unphysical φ_s range: [{sol.φ_s.min():.2f}, {sol.φ_s.max():.2f}]")
        return False

    return True



def physical_initial_guess(params: Parameters, mesh: Mesh, V_cell: float) -> np.ndarray:
    """Generate realistic initial guess."""
    n = mesh.n
    
    E_rev_anode = 1.478 - 8.347e-4 * params.T  # ~1.19V at 353K
    E_rev_cathode = 0.0
    
    # At 1.8V cell: need ~0.6V anode overpotential for OER
    # Estimate based on Tafel: at 1 A/cm², η_a ≈ b*log(j/j0)
    if V_cell < 1.6:
        η_anode = 0.3
    elif V_cell < 1.9:
        η_anode = 0.4 + (V_cell - 1.6) * 0.5  # Scale with voltage
    else:
        η_anode = 0.6
    
    η_cathode = 0.05  # HER is fast
    V_ohmic = max(0.1, V_cell - E_rev_anode - η_anode - η_cathode)
    
    φ_s = np.zeros(n)
    φ_m = np.zeros(n)
    
    # Electronic potential (anode = V_cell, cathode = 0)
    φ_s[mesh.idx_PTL] = V_cell
    φ_s[mesh.idx_aCL] = V_cell - 0.02  # Small drop in aCL
    φ_s[mesh.idx_mem] = E_rev_anode + η_anode  # Transition through membrane
    φ_s[mesh.idx_cCL] = η_cathode * 0.5  # Near zero
    φ_s[mesh.idx_GDL] = 0.0
    
    # Ionic potential (drops across membrane)
    φ_m[mesh.idx_PTL] = 0.0
    
    # aCL: φ_m rises as reaction proceeds
    for i in mesh.idx_aCL:
        progress = (i - mesh.idx_aCL.start) / max(1, len(mesh.idx_aCL) - 1)
        # At aCL: φ_s - φ_m = E_rev + η
        φ_m[i] = φ_s[i] - E_rev_anode - η_anode
    
    # Membrane: ohmic drop
    for i in mesh.idx_mem:
        progress = (i - mesh.idx_mem.start) / max(1, len(mesh.idx_mem) - 1)
        φ_m[i] = φ_m[mesh.idx_aCL.stop - 1] - V_ohmic * progress
    
    # cCL: φ_m continues dropping
    for i in mesh.idx_cCL:
        progress = (i - mesh.idx_cCL.start) / max(1, len(mesh.idx_cCL) - 1)
        φ_m[i] = φ_m[mesh.idx_mem.stop - 1] - η_cathode * progress
    
    φ_m[mesh.idx_GDL] = φ_m[mesh.idx_cCL.stop - 1]
    
    # Gauge
    i_gauge = mesh.idx_mem.start + len(mesh.idx_mem)//2
    φ_m -= φ_m[i_gauge]
    
    # Concentrations (keep your existing code)
    c_H2_cath = params.P_cathode * params.S_H2
    c_O2_an = params.P_anode * params.S_O2
    c_H2 = np.linspace(0.1 * c_H2_cath, c_H2_cath, n)
    c_O2 = np.linspace(c_O2_an, 0.1 * c_O2_an, n)
    
    return np.concatenate([φ_s, φ_m, c_H2, c_O2])


def residual(u: np.ndarray, params: Parameters, mesh: Mesh,
            A_s: csr_matrix, A_m: csr_matrix,
            A_H2: csr_matrix, A_O2: csr_matrix, 
            V_cell: float, alpha: float = 1.0) -> np.ndarray:
    """Compute residual for the coupled system."""
    n = mesh.n
    
    # Unpack state
    φ_s = u[:n]
    φ_m = u[n:2*n] 
    c_H2 = u[2*n:3*n]
    c_O2 = u[3*n:4*n]
    
    # Current density (scaled by alpha for homotopy)
    I = current_density(params, mesh, φ_s, φ_m, c_H2, c_O2, alpha=alpha)
    
    # Initialize residual
    res = np.zeros(4*n)
    
    # Charge balances
    res[:n] = A_s.dot(φ_s) - I * mesh.Δz
    res[n:2*n] = A_m.dot(φ_m) + I * mesh.Δz
    
    # Volume fractions for species equations
    _, ε_ion, _, ε_H2O_ion = compute_volume_fractions(params, mesh)
    pref = np.zeros(n)
    for i in (*mesh.idx_aCL, *mesh.idx_mem, *mesh.idx_cCL):
        pref[i] = 1.0 / max(ε_ion[i] * ε_H2O_ion[i], 1e-12)
    
    # Diffusion terms
    diff_H2 = A_H2.dot(c_H2)
    diff_O2 = A_O2.dot(c_O2)
    res[2*n:3*n] = pref * diff_H2
    res[3*n:4*n] = pref * diff_O2
    
    # Electro-osmotic drag advection
    σP_eff = np.zeros(n)
    σm_cell = params.σ_m_ref * np.exp(1268*(1/353.0 - 1/params.T))
    for i in (*mesh.idx_aCL, *mesh.idx_mem, *mesh.idx_cCL):
        σP_eff[i] = σm_cell * (ε_ion[i]**1.5 if i not in mesh.idx_mem else 1.0)
    
    c_H2O_molar = 54000.0  # mol/m³
    
    for i in range(1, n-1):
        if pref[i] == 0:
            continue
            
        # Face transmissibilities
        T_e = 0.0
        if σP_eff[i] > 0 and σP_eff[i+1] > 0:
            T_e = 1.0 / ((mesh.Δz[i]*0.5)/σP_eff[i] + (mesh.Δz[i+1]*0.5)/σP_eff[i+1])
        T_w = 0.0  
        if σP_eff[i-1] > 0 and σP_eff[i] > 0:
            T_w = 1.0 / ((mesh.Δz[i]*0.5)/σP_eff[i] + (mesh.Δz[i-1]*0.5)/σP_eff[i-1])
        
        # Proton current at faces
        j_e = -T_e * (φ_m[i+1] - φ_m[i])
        j_w = -T_w * (φ_m[i] - φ_m[i-1])
        
        # Drag velocities
        v_e = params.n_drag * j_e / (F * c_H2O_molar)
        v_w = params.n_drag * j_w / (F * c_H2O_molar)
        
        # Upwind advection
        flux_e_H2 = v_e * (c_H2[i] if v_e >= 0 else c_H2[i+1])
        flux_w_H2 = v_w * (c_H2[i-1] if v_w >= 0 else c_H2[i])
        flux_e_O2 = v_e * (c_O2[i] if v_e >= 0 else c_O2[i+1])
        flux_w_O2 = v_w * (c_O2[i-1] if v_w >= 0 else c_O2[i])
        
        res[2*n + i] += pref[i] * (flux_e_H2 - flux_w_H2) / mesh.Δz[i]
        res[3*n + i] += pref[i] * (flux_e_O2 - flux_w_O2) / mesh.Δz[i]
    
    # Mass transfer and production in catalyst layers
    for i in mesh.idx_cCL:
        c_sat = params.P_cathode * params.S_H2
        Γ_H2_prod = abs(I[i]) / (2*F)
        Γ_H2_mass = params.a_interface_cCL * params.k_H2_mass * (c_H2[i] - c_sat)
        Γ_O2_mass = params.a_interface_cCL * params.k_O2_mass * c_O2[i]
        res[2*n + i] += pref[i] * (Γ_H2_prod - Γ_H2_mass) * mesh.Δz[i]
        res[3*n + i] += pref[i] * (-Γ_O2_mass) * mesh.Δz[i]
        
    for i in mesh.idx_aCL:
        c_sat = params.P_anode * params.S_O2
        Γ_O2_prod = I[i] / (4*F)
        Γ_O2_mass = params.a_interface_aCL * params.k_O2_mass * (c_O2[i] - c_sat)
        Γ_H2_mass = params.a_interface_aCL * params.k_H2_mass * c_H2[i]
        res[3*n + i] += pref[i] * (Γ_O2_prod - Γ_O2_mass) * mesh.Δz[i]
        res[2*n + i] += pref[i] * (-Γ_H2_mass) * mesh.Δz[i]
    
    # Boundary conditions
    # Potential BCs
    res[0] = φ_s[0] - V_cell
    res[n-1] = φ_s[n-1]
    
    # Proton gauge
    i_gauge = mesh.idx_mem.start + len(mesh.idx_mem)//2
    res[n + i_gauge] = φ_m[i_gauge]
    
    # Species BCs: zero-gradient at boundaries
    i_ptl_last = mesh.idx_PTL.stop - 1
    i_gdl_first = mesh.idx_GDL.start
    
    res[2*n + 0] = c_H2[1] - c_H2[0]
    res[2*n + n-1] = c_H2[n-1] - c_H2[n-2]
    res[2*n + i_ptl_last] = c_H2[i_ptl_last] - c_H2[i_ptl_last + 1]
    res[2*n + i_gdl_first] = c_H2[i_gdl_first] - c_H2[i_gdl_first - 1]
    
    res[3*n + 0] = c_O2[1] - c_O2[0]
    res[3*n + n-1] = c_O2[n-1] - c_O2[n-2]
    res[3*n + i_ptl_last] = c_O2[i_ptl_last] - c_O2[i_ptl_last + 1]
    res[3*n + i_gdl_first] = c_O2[i_gdl_first] - c_O2[i_gdl_first - 1]
    
    # Scale residual for better conditioning
    scale_V = 1.0 / max(V_cell, 1.0)
    scale_c = 1.0 / 100.0
    
    res[:n] *= scale_V
    res[n:2*n] *= scale_V
    res[2*n:3*n] *= scale_c
    res[3*n:4*n] *= scale_c
    
    return res


def precompute_operators(params: Parameters, mesh: Mesh):
    """Precompute finite-volume matrices and fields."""
    σ_s, σ_m = effective_conductivities(params, mesh)
    A_s = build_conduction_matrix(mesh, σ_s)
    A_m = build_conduction_matrix(mesh, σ_m)

    D_H2, D_O2 = effective_diffusion(params, mesh)
    A_H2 = build_diffusion_matrix(mesh, D_H2)
    A_O2 = build_diffusion_matrix(mesh, D_O2)
    
    return {
        'A_s': A_s, 'A_m': A_m, 'A_H2': A_H2, 'A_O2': A_O2,
        'σ_s': σ_s, 'σ_m': σ_m, 'D_H2': D_H2, 'D_O2': D_O2
    }


def solve_steady_state(params: Parameters, mesh: Mesh, V_cell: float = 1.8,
                       previous_sol: Optional[Solution] = None, 
                       tol: float = 1e-6, max_iter: int = 5000,
                       verbose: bool = False, ops: Optional[dict] = None) -> Solution:
    """Solve steady-state with improved homotopy and physical checking."""
    n = mesh.n
    
    # Precompute operators if needed
    if ops is None:
        ops = precompute_operators(params, mesh)
    
    # Initial guess
    if previous_sol is None:
        u0 = physical_initial_guess(params, mesh, V_cell)
    else:
        # Warm start from previous solution
        ΔV = V_cell - (previous_sol.φ_s[0] - previous_sol.φ_s[-1])
        ramp = np.linspace(1.0, 0.0, n)
        u0 = np.concatenate([
            previous_sol.φ_s + ΔV * ramp * 0.8,  # Scale adjustment
            previous_sol.φ_m,
            previous_sol.c_H2,
            previous_sol.c_O2
        ])
    
    # Improved homotopy with finer steps
    u = u0.copy()
    nfev_total = 0
    success = False
    
    # Use finer homotopy steps as recommended
    if previous_sol is None:
        alpha_steps = [1e-3, 3e-3, 1e-2, 3e-2, 1e-1, 2e-1, 4e-1, 6e-1, 8e-1, 1.0]
    else:
        alpha_steps = [0.5, 1.0]  # Less aggressive for warm starts
    
    for α in alpha_steps:
        def f(x):
            return residual(x, params, mesh, ops['A_s'], ops['A_m'], 
                          ops['A_H2'], ops['A_O2'], V_cell, alpha=α)
        
        # Solve with current alpha
        sol = root(f, u, method='hybr', tol=tol, options={'maxfev': max_iter//len(alpha_steps)})
        nfev_total += sol.nfev
        
        # Check if solution is physical
        u_test = sol.x
        φ_s_test = u_test[:n]
        φ_m_test = u_test[n:2*n]
        c_H2_test = np.clip(u_test[2*n:3*n], 0, np.inf)
        c_O2_test = np.clip(u_test[3*n:4*n], 0, np.inf)
        
        I_test = current_density(params, mesh, φ_s_test, φ_m_test, c_H2_test, c_O2_test, alpha=α)
        sol_test = Solution(φ_s_test, φ_m_test, c_H2_test, c_O2_test, I_test, sol.success, nfev_total)
        
        if sol.success and check_physical_solution(sol_test, mesh, verbose=verbose):
            u = sol.x
            success = True
            if verbose and α < 1.0:
                print(f"  α={α:.3f}: converged, continuing...")
        else:
            if verbose:
                print(f"  α={α:.3f}: failed or unphysical, trying smaller α...")
            # Try with smaller step
            if α > 1e-3:
                # Reduce alpha and retry from previous good solution
                alpha_steps = [α/3, α/2] + [a for a in alpha_steps if a > α]
            else:
                break
    
    # Extract final solution
    φ_s = u[:n]
    φ_m = u[n:2*n]
    c_H2 = np.clip(u[2*n:3*n], 0, np.inf)
    c_O2 = np.clip(u[3*n:4*n], 0, np.inf)
    
    I = current_density(params, mesh, φ_s, φ_m, c_H2, c_O2, alpha=1.0)
    
    # Final physical check
    final_sol = Solution(φ_s, φ_m, c_H2, c_O2, I, success, nfev_total)
    if not check_physical_solution(final_sol, mesh, verbose=False):
        success = False
    
    if verbose:
        j_total = np.sum(I * mesh.Δz) / 1e4
        print(f"V={V_cell:.2f}: j={j_total:.4f} A/cm², converged={success}, iters={nfev_total}")
    
    return final_sol


def calculate_current_density(sol: Solution, mesh: Mesh, σ_s: np.ndarray = None) -> float:
    """Return anodic current density (A/cm^2), positive by convention."""
    # Hard fail if any non-finite appears
    for arr in (sol.φ_s, sol.φ_m, sol.I):
        if not np.all(np.isfinite(arr)):
            return float('nan')

    if σ_s is not None and sol.converged:
        i = mesh.idx_PTL.stop - 2          # last PTL node (left of aCL)
        j = i + 1                           # first aCL node
        if σ_s[i] > 0 and σ_s[j] > 0:
            Te = 1.0 / ((mesh.Δz[i]*0.5)/σ_s[i] + (mesh.Δz[j]*0.5)/σ_s[j])
            # Positive when φ_s[left] > φ_s[right] (anodic current into cell)
            j_s = Te * (sol.φ_s[i] - sol.φ_s[j])   # A/m^2
            return j_s / 1e4
    # Fallback (should also be positive with conventions)
    j_a = float(np.sum(sol.I[mesh.idx_aCL] * mesh.Δz[mesh.idx_aCL]))
    return j_a / 1e4



def voltage_from_current(params: Parameters, mesh: Mesh, j_target: float,
                         V_guess: float = 1.8, tol: float = 1e-4,
                         target_unit: str = "A/cm2", verbose: bool = False) -> float:
    """Find V_cell that yields the desired current density."""
    # Convert target to A/cm² if needed
    if target_unit.lower() in ("a/m2", "a_per_m2", "si"):
        j_target = j_target / 1e4
    
    # Bracket the solution
    V_low, V_high = 1.4, 2.5
    ops = precompute_operators(params, mesh)
    
    # Cache for warm starts
    last_sol = None
    
    def j_at(V):
        nonlocal last_sol
        sol = solve_steady_state(params, mesh, V_cell=V, verbose=False, ops=ops,
                                  previous_sol=last_sol)
        if not sol.converged:
            last_sol = None
            return float('nan')
        last_sol = sol
        return calculate_current_density(sol, mesh, ops['σ_s'])
    
    # Find initial bracket
    j_low = j_at(V_low)
    j_high = j_at(V_high)
    
    tries = 0
    while not (np.isfinite(j_low) and np.isfinite(j_high) and j_low <= j_target <= j_high) and tries < 20:
        if (not np.isfinite(j_low)) or (j_low > j_target):
            V_low = max(0.5, V_low - 0.1)
        if (not np.isfinite(j_high)) or (j_high < j_target):
            V_high = V_high + 0.1
        j_low, j_high = j_at(V_low), j_at(V_high)
        tries += 1
    
    if not (np.isfinite(j_low) and np.isfinite(j_high) and j_low <= j_target <= j_high):
        if verbose:
            print(f"Could not bracket j={j_target} A/cm²")
        return V_guess
    
    # Bisection
    for _ in range(30):
        V_mid = 0.5*(V_low + V_high)
        j_mid = j_at(V_mid)
        if not np.isfinite(j_mid):
            V_low -= 0.05
            V_high += 0.05
            continue
        if abs(j_mid - j_target) < tol:
            return V_mid
        if j_mid < j_target:
            V_low = V_mid
        else:
            V_high = V_mid
    
    return 0.5*(V_low + V_high)


# Include the transient solver and other functions...
def solve_potentials_quasi_steady_inplace(
    φs: np.ndarray, φm: np.ndarray,
    params: Parameters, mesh: Mesh,
    c_H2: np.ndarray, c_O2: np.ndarray,
    V_cell: float,
    A_s: csr_matrix, A_m: csr_matrix) -> None:
    """Solve for quasi-steady potentials in-place."""
    n = mesh.n
    u = np.concatenate([φs, φm])
    i_gauge = mesh.idx_mem.start + len(mesh.idx_mem)//2

    def res_pot(u_pot):
        φs_vec = u_pot[:n]
        φm_vec = u_pot[n:2*n]
        I_local = current_density(params, mesh, φs_vec, φm_vec, c_H2, c_O2, alpha=1.0)

        res = np.zeros(2*n)
        res[:n] = A_s.dot(φs_vec) - I_local * mesh.Δz
        res[n:2*n] = A_m.dot(φm_vec) + I_local * mesh.Δz
        res[0] = φs_vec[0] - V_cell
        res[n-1] = φs_vec[n-1] - 0.0
        res[n + i_gauge] = φm_vec[i_gauge] - 0.0

        return res

    sol = root(res_pot, u, method='hybr', tol=1e-8, options={'maxfev': 100})
    φs[:] = sol.x[:n]
    φm[:] = sol.x[n:2*n]


@dataclass
class TransientSolution:
    t: np.ndarray
    H2_in_O2_channel: np.ndarray
    H2_in_O2_aCL: np.ndarray
    c_H2_profiles: np.ndarray
    c_O2_profiles: np.ndarray
    j_applied: np.ndarray
    j_measured: np.ndarray  
    V_trace: np.ndarray


def solve_transient(params: Parameters, mesh: Mesh,
                   j_initial: float = 4.5,
                   j_final: float = 1.62,
                   t_step: float = 50.0,
                   t_span: Tuple[float, float] = (0.0, 60.0),
                   verbose: bool = False) -> TransientSolution:
    """Solve the transient problem for a current density step change."""
    n = mesh.n
    
    # Get initial and final voltages
    if verbose:
        print("Finding initial voltage...")
    V_initial = voltage_from_current(params, mesh, j_initial, verbose=verbose)
    sol_initial = solve_steady_state(params, mesh, V_cell=V_initial, verbose=verbose)
    
    if verbose:
        print("Finding final voltage...")
    V_final = voltage_from_current(params, mesh, j_final, verbose=verbose)
    
    # Initial conditions
    u0 = np.concatenate([sol_initial.c_H2, sol_initial.c_O2])
    
    # Precompute constants
    σ_s, σ_m = effective_conductivities(params, mesh)
    A_s = build_conduction_matrix(mesh, σ_s)
    A_m = build_conduction_matrix(mesh, σ_m)
    D_H2, D_O2 = effective_diffusion(params, mesh)
    
    # Volume fractions
    _, ε_ion, _, ε_H2O_ion = compute_volume_fractions(params, mesh)
    pref = np.zeros(n)
    diss_idxs = list(range(mesh.idx_aCL.start, mesh.idx_aCL.stop)) + \
                list(range(mesh.idx_mem.start, mesh.idx_mem.stop)) + \
                list(range(mesh.idx_cCL.start, mesh.idx_cCL.stop))
    for i in diss_idxs:
        pref[i] = 1.0 / max(ε_ion[i] * ε_H2O_ion[i], 1e-12)
    
    # Effective proton conductivity
    σP_eff = np.zeros(n)
    σP_eff_base = protonic_conductivity_at_T(params.λ, params.T)
    for i in diss_idxs:
        σP_eff[i] = σP_eff_base * (ε_ion[i]**1.5 if i not in mesh.idx_mem else 1.0)
    
    # Saturation concentrations
    c_H2_sat_cath = params.P_cathode * params.S_H2
    c_O2_sat_an = params.P_anode * params.S_O2
    
    # Warm-start caches
    φs_cache = sol_initial.φ_s.copy()
    φm_cache = sol_initial.φ_m.copy()
    
    def dudt(t, u):
        c_H2 = u[:n]
        c_O2 = u[n:2*n]
        dc_H2_dt = np.zeros(n)
        dc_O2_dt = np.zeros(n)
        
        # Target current
        j_set = j_initial if t < t_step else j_final
        
        # Galvanostatic control
        def j_for_voltage(V_try, φs_guess, φm_guess):
            solve_potentials_quasi_steady_inplace(
                φs_guess, φm_guess, params, mesh, c_H2, c_O2, V_try, A_s, A_m
            )
            I_try = current_density(params, mesh, φs_guess, φm_guess, c_H2, c_O2)
            return float(np.sum(I_try * mesh.Δz)) / 1e4
        
        # Find voltage for target current
        V0 = getattr(dudt, "_V_last", V_initial if t < t_step else V_final)
        φs0, φm0 = φs_cache.copy(), φm_cache.copy()
        j0 = j_for_voltage(V0, φs0, φm0)
        
        # Simple iteration
        for _ in range(5):
            if abs(j0 - j_set) < 1e-4:
                break
            V1 = V0 + (j_set - j0) * 0.05  # Small adjustment
            V1 = np.clip(V1, 1.2, 2.8)
            φs1, φm1 = φs_cache.copy(), φm_cache.copy()
            j1 = j_for_voltage(V1, φs1, φm1)
            V0, j0 = V1, j1
        
        V_cell = V0
        dudt._V_last = V_cell
        
        # Update cached potentials
        solve_potentials_quasi_steady_inplace(
            φs_cache, φm_cache, params, mesh, c_H2, c_O2, V_cell, A_s, A_m
        )
        I_loc = current_density(params, mesh, φs_cache, φm_cache, c_H2, c_O2)
        
        # Transport equations (simplified for brevity - full version similar to original)
        dz = mesh.Δz

        # Diffusion matrices (paper equation 15)
        for i in range(1, n-1):
            if pref[i] == 0:
                continue
            
            # Diffusion transmissibilities
            D_H2_eff = D_H2  # from effective_diffusion()
            D_O2_eff = D_O2
            
            # East/west transmissibilities
            T_e_H2 = 0.0
            if D_H2_eff[i] > 0 and D_H2_eff[i+1] > 0:
                T_e_H2 = 1.0 / ((dz[i]*0.5)/D_H2_eff[i] + (dz[i+1]*0.5)/D_H2_eff[i+1])
            T_w_H2 = 0.0
            if D_H2_eff[i-1] > 0 and D_H2_eff[i] > 0:
                T_w_H2 = 1.0 / ((dz[i]*0.5)/D_H2_eff[i] + (dz[i-1]*0.5)/D_H2_eff[i-1])
            
            # Diffusion term
            dc_H2_dt[i] += pref[i] * (T_e_H2 * (c_H2[i+1] - c_H2[i]) - T_w_H2 * (c_H2[i] - c_H2[i-1])) / dz[i]
            
            # Same for O2
            T_e_O2 = 0.0
            if D_O2_eff[i] > 0 and D_O2_eff[i+1] > 0:
                T_e_O2 = 1.0 / ((dz[i]*0.5)/D_O2_eff[i] + (dz[i+1]*0.5)/D_O2_eff[i+1])
            T_w_O2 = 0.0
            if D_O2_eff[i-1] > 0 and D_O2_eff[i] > 0:
                T_w_O2 = 1.0 / ((dz[i]*0.5)/D_O2_eff[i] + (dz[i-1]*0.5)/D_O2_eff[i-1])
            
            dc_O2_dt[i] += pref[i] * (T_e_O2 * (c_O2[i+1] - c_O2[i]) - T_w_O2 * (c_O2[i] - c_O2[i-1])) / dz[i]
            
            # Electro-osmotic drag advection (paper equation 15)
            # Calculate proton current at faces
            j_p_e = -σP_eff[i] * (φm_cache[i+1] - φm_cache[i]) / (0.5 * (dz[i] + dz[i+1]))
            j_p_w = -σP_eff[i-1] * (φm_cache[i] - φm_cache[i-1]) / (0.5 * (dz[i] + dz[i-1]))
            
            # Drag velocity: v = n_drag * j_p / (F * c_H2O)
            c_H2O_molar = 54000.0  # mol/m³
            v_drag_e = params.n_drag * j_p_e / (F * c_H2O_molar)
            v_drag_w = params.n_drag * j_p_w / (F * c_H2O_molar)
            
            # Upwind advection
            flux_e_H2 = v_drag_e * (c_H2[i] if v_drag_e >= 0 else c_H2[i+1])
            flux_w_H2 = v_drag_w * (c_H2[i-1] if v_drag_w >= 0 else c_H2[i])
            dc_H2_dt[i] += pref[i] * (flux_e_H2 - flux_w_H2) / dz[i]
            
            flux_e_O2 = v_drag_e * (c_O2[i] if v_drag_e >= 0 else c_O2[i+1])
            flux_w_O2 = v_drag_w * (c_O2[i-1] if v_drag_w >= 0 else c_O2[i])
            dc_O2_dt[i] += pref[i] * (flux_e_O2 - flux_w_O2) / dz[i]
        

        
        # Add reaction sources
        ia = slice(mesh.idx_aCL.start, mesh.idx_aCL.stop)
        ic = slice(mesh.idx_cCL.start, mesh.idx_cCL.stop)
        
        # Cathode HER
        Γ_H2_prod_c = np.abs(I_loc[ic]) / (2.0 * F)
        Γ_H2_mass_c = params.a_interface_cCL * params.k_H2_mass * (c_H2[ic] - c_H2_sat_cath)
        Γ_O2_mass_c = params.a_interface_cCL * params.k_O2_mass * c_O2[ic]
        dc_H2_dt[ic] += pref[ic] * (Γ_H2_prod_c - Γ_H2_mass_c)
        dc_O2_dt[ic] += pref[ic] * (-Γ_O2_mass_c)
        
        # Anode OER
        Γ_O2_prod_a = I_loc[ia] / (4.0 * F)
        Γ_O2_mass_a = params.a_interface_aCL * params.k_O2_mass * (c_O2[ia] - c_O2_sat_an)
        Γ_H2_mass_a = params.a_interface_aCL * params.k_H2_mass * c_H2[ia]
        dc_O2_dt[ia] += pref[ia] * (Γ_O2_prod_a - Γ_O2_mass_a)
        dc_H2_dt[ia] += pref[ia] * (-Γ_H2_mass_a)
        
        # No-flux boundaries
        dc_H2_dt[[0, -1]] = 0.0
        dc_O2_dt[[0, -1]] = 0.0
        
        return np.concatenate([dc_H2_dt, dc_O2_dt])
    
    # Integrate
    t_eval = np.arange(t_span[0], t_span[1] + 0.1, 0.1)
    sol_ode = solve_ivp(dudt, t_span, u0, t_eval=t_eval, method='Radau',
                       atol=1e-8, rtol=1e-6)
    if not sol_ode.success:
        print("[transient] IVP failed:", sol_ode.message)
    
    # Post-process results
    n_times = len(sol_ode.t)
    H2_in_O2_channel = np.zeros(n_times)
    H2_in_O2_aCL = np.zeros(n_times)
    c_H2_profiles = np.zeros((n, n_times))
    c_O2_profiles = np.zeros((n, n_times))
    j_applied = np.zeros(n_times)
    j_measured = np.zeros(n_times)
    V_trace = np.zeros(n_times)
    
    for k, tcur in enumerate(sol_ode.t):
        c_H2_k = sol_ode.y[:n, k]
        c_O2_k = sol_ode.y[n:2*n, k]
        c_H2_profiles[:, k] = c_H2_k
        c_O2_profiles[:, k] = c_O2_k
        j_applied[k] = j_initial if tcur < t_step else j_final
        
        # Calculate H2-in-O2 fractions

        i_anode_boundary = mesh.idx_PTL.start  # leftmost cell (anode gas side)
        c_H2_b = max(c_H2_k[i_anode_boundary], 0.0)
        c_O2_b = max(c_O2_k[i_anode_boundary], 0.0)
        y_H2_b = c_H2_b / max(params.S_H2, 1e-30)
        y_O2_b = c_O2_b / max(params.S_O2, 1e-30)
        den = y_H2_b + y_O2_b
        H2_in_O2_channel[k] = 100.0 * (y_H2_b / den if den > 0 else 0.0)


        
        i_aCL_mem = mesh.idx_aCL.stop - 1
        local_H2 = c_H2_k[i_aCL_mem] / params.S_H2
        local_O2 = c_O2_k[i_aCL_mem] / params.S_O2
        H2_in_O2_aCL[k] = 100.0 * local_H2 / (local_O2 + local_H2 + 1e-30)
    
    return TransientSolution(sol_ode.t, H2_in_O2_channel, H2_in_O2_aCL,
                             c_H2_profiles, c_O2_profiles, j_applied,
                             j_measured, V_trace)


def main(voltages=None,
         j_initial: float = 4.5,
         j_final: float = 1.62,
         t_step: float = 50.0,
         t_span: Tuple[float, float] = (0.0, 60.0),
         T: float = 353.0,
         membrane: str = 'N117',
         pressure: float = 1e5,
         k_H2_mass_value: float = 540.0,
         k_O2_mass_value: Optional[float] = None,
         do_plot: bool = True):
    """Run steady polarization and galvanostatic transient analysis."""
    
    if voltages is None:
        voltages = [1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.1, 2.2]
    
    # Use coarser mesh initially for better convergence
    params = default_parameters(T=T, membrane=membrane, pressure=pressure)
    mesh = create_mesh(params, n_PTL=10, n_aCL=10, n_mem=10, n_cCL=10, n_GDL=10)
    
    print("=== STEADY-STATE POLARIZATION CURVE ===\n")
    print(f"{'Voltage[V]':<12} | {'Current[A/cm²]':<20} | {'Converged':<10} | {'Iters':<10}")
    print("-" * 65)
    
    currents, iters, convs = [], [], []
    prev_sol = None
    ops = precompute_operators(params, mesh)
    
    for V_cell in voltages:
        sol = solve_steady_state(params, mesh, V_cell=V_cell,
                                 previous_sol=prev_sol, tol=1e-6, max_iter=5000,
                                 verbose=False, ops=ops)
        
        if sol.converged and check_physical_solution(sol, mesh):
            j = calculate_current_density(sol, mesh, ops['σ_s'])
            prev_sol = sol
        else:
            j = float('nan')
            prev_sol = None
        
        currents.append(j)
        iters.append(sol.iterations)
        convs.append(sol.converged)
        print(f"{V_cell:<12.2f} | {j:<20.4f} | {str(sol.converged):<10} | {sol.iterations:<10d}")
    
    # Transient analysis
    print(f"\n=== TRANSIENT ANALYSIS ===")
    print(f"Simulating current step: {j_initial} → {j_final} A/cm² at t = {t_step} s")
    
    p_tr = default_parameters(T=T, membrane=membrane, pressure=pressure)
    p_tr.k_H2_mass = k_H2_mass_value
    if k_O2_mass_value is not None:
        p_tr.k_O2_mass = k_O2_mass_value
    
    tr = solve_transient(p_tr, mesh,
                         j_initial=j_initial, j_final=j_final,
                         t_step=t_step, t_span=t_span, verbose=True)
    
    # Plotting
    plot_path = ""
    if do_plot:
        plt.figure(figsize=(14, 10))
        
        # Polarization curve
        plt.subplot(2, 2, 1)
        valid_idx = np.isfinite(currents)
        if np.any(valid_idx):
            plt.plot(np.array(voltages)[valid_idx], np.array(currents)[valid_idx], 'o-', lw=2)
            plt.xlabel("Cell Voltage (V)")
            plt.ylabel("Current Density (A/cm²)")
            plt.title("Polarization Curve")
            plt.grid(True, alpha=0.3)
        
        # H2-in-O2 transient
        plt.subplot(2, 2, 2)
        plt.plot(tr.t, tr.H2_in_O2_channel, label="H₂-in-O₂ (channel)", lw=2)
        plt.plot(tr.t, tr.H2_in_O2_aCL, '--', label="H₂-in-O₂ (aCL/mem)", lw=2)
        plt.axhline(y=4.0, linestyle=':', color='red', lw=1, label="LEL (4%)")
        plt.axhline(y=2.0, linestyle=':', color='orange', lw=1, label="50% LEL")
        plt.axvline(x=t_step, linestyle='--', color='gray', lw=1, label="Current step")
        plt.xlabel("Time (s)")
        plt.ylabel("H₂-in-O₂ (vol %)")
        plt.title(f"Transient H₂ Crossover")
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Applied current
        plt.subplot(2, 2, 3)
        plt.plot(tr.t, tr.j_applied, lw=2)
        plt.xlabel("Time (s)")
        plt.ylabel("Current Density (A/cm²)")
        plt.title("Applied Current Density")
        plt.grid(True, alpha=0.3)
        
        # Concentration profiles at key times
        plt.subplot(2, 2, 4)
        times_to_plot = [0, t_step-1, t_step+1, t_span[1]]
        for t_plot in times_to_plot:
            idx = np.argmin(np.abs(tr.t - t_plot))
            plt.plot(mesh.z*1e6, tr.c_H2_profiles[:, idx], 
                    label=f"t={tr.t[idx]:.1f}s", alpha=0.7)
        plt.xlabel("Position (μm)")
        plt.ylabel("H₂ Concentration (mol/m³)")
        plt.title("H₂ Concentration Profiles")
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plot_path = f"pem_electrolyzer_analysis_{membrane}_{int(T)}K.png"
        plt.savefig(plot_path, dpi=150)
        plt.show()
        print(f"\nPlot saved: {plot_path}")
    
    # Summary
    max_H2 = float(np.max(tr.H2_in_O2_channel))
    steady_initial = float(tr.H2_in_O2_channel[0])
    steady_final = float(tr.H2_in_O2_channel[-1])
    
    print("\n=== TRANSIENT SUMMARY ===")
    print(f"Initial steady-state H₂-in-O₂: {steady_initial:.3f} vol%")
    print(f"Peak H₂-in-O₂ (overshoot):     {max_H2:.3f} vol%")
    print(f"Final steady-state H₂-in-O₂:   {steady_final:.3f} vol%")
    print(f"Overshoot magnitude:            {max_H2 - steady_final:.3f} vol%")
    
    if max_H2 > 4.0:
        print("\n⚠️  WARNING: Peak exceeds Lower Explosion Limit (4%)!")
    elif max_H2 > 2.0:
        print("\n⚠️  WARNING: Peak exceeds 50% of LEL (2%)!")
    
    return {
        'voltages': voltages,
        'currents': currents,
        'iterations': iters,
        'converged': convs,
        'transient': tr,
        'plot_path': plot_path,
        'params': params,
        'mesh': mesh,
        'ops': ops,
    }


if __name__ == "__main__":
    results = main(
        voltages=[1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.1, 2.2],
        j_initial=4.5,
        j_final=1.62,
        t_step=50.0,
        t_span=(0.0, 60.0),
        T=353.0,
        membrane='N117',
        pressure=1e5,
        k_H2_mass_value=540.0,
        do_plot=True
    )


# =========================
# DEBUG REPORT: SAVE INPUTS, DERIVEDS, & RESULTS
# =========================
from dataclasses import asdict
from datetime import datetime
import numpy as np
import os
import platform
import sys

def _fmt_range(r: range) -> str:
    return f"[start={r.start}, stop={r.stop}, len={len(r)}]"

def _summ_stats(arr: np.ndarray, name: str, units: str = "", values_if_small: bool = True) -> str:
    a = np.asarray(arr)
    flat = a.ravel()
    stats = {
        "shape": a.shape,
        "dtype": str(a.dtype),
        "min": float(np.nanmin(flat)) if flat.size else float("nan"),
        "max": float(np.nanmax(flat)) if flat.size else float("nan"),
        "mean": float(np.nanmean(flat)) if flat.size else float("nan"),
        "std": float(np.nanstd(flat)) if flat.size else float("nan"),
    }
    s = []
    s.append(f"{name} {units}".strip())
    s.append(f"  shape={stats['shape']} dtype={stats['dtype']}")
    s.append(f"  min={stats['min']:.6g}  max={stats['max']:.6g}  mean={stats['mean']:.6g}  std={stats['std']:.6g}")
    # If reasonably small, print all values (one line per 10 elements)
    if values_if_small and flat.size <= 200:
        s.append("  values:")
        line = []
        for i, v in enumerate(flat):
            line.append(f"{v:.6g}")
            if (i+1) % 10 == 0 or i == flat.size - 1:
                s.append("    " + ", ".join(line))
                line = []
    return "\n".join(s)

def write_debug_report(results: dict, filename: str | None = None) -> str:
    """Write a human-readable debug report capturing inputs and calculated parameters."""
    if filename is None:
        stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"pem_debug_report_{stamp}.txt"

    lines = []
    lines.append("="*78)
    lines.append("PEM Electrolyzer Debug Report")
    lines.append("="*78)
    lines.append(f"Created: {datetime.now().isoformat(timespec='seconds')}")
    lines.append(f"Python : {sys.version.split()[0]}  |  NumPy: {np.__version__}")
    lines.append(f"Platform: {platform.platform()}")
    lines.append("")

    # ---- High-level run configuration
    lines.append("# RUN CONFIG")
    for k in ("voltages", "currents", "iterations", "converged"):
        if k in results:
            v = results[k]
            lines.append(f"{k}: {v}")
    lines.append("")

    # ---- Parameters (inputs and temperature-corrected values living in the dataclass)
    params = results.get("params", None)
    if params is not None:
        lines.append("# PARAMETERS (dataclass fields)")
        pd = asdict(params)
        # sort keys for consistency
        for k in sorted(pd.keys()):
            v = pd[k]
            if isinstance(v, float):
                lines.append(f"{k:>20s}: {v:.10g}")
            else:
                lines.append(f"{k:>20s}: {v}")
        lines.append("")

    # ---- Mesh summary
    mesh = results.get("mesh", None)
    if mesh is not None:
        lines.append("# MESH")
        lines.append(f"n={mesh.n}")
        lines.append(f"Δz stats: min={float(np.min(mesh.Δz)):.6g} m, max={float(np.max(mesh.Δz)):.6g} m")
        lines.append(f"idx_PTL: {_fmt_range(mesh.idx_PTL)}")
        lines.append(f"idx_aCL: {_fmt_range(mesh.idx_aCL)}")
        lines.append(f"idx_mem: {_fmt_range(mesh.idx_mem)}")
        lines.append(f"idx_cCL: {_fmt_range(mesh.idx_cCL)}")
        lines.append(f"idx_GDL: {_fmt_range(mesh.idx_GDL)}")
        lines.append(_summ_stats(mesh.z, "z (cell centers)", " [m]"))
        lines.append(_summ_stats(mesh.Δz, "Δz", " [m]"))
        lines.append("")

    # ---- Precomputed operators & effective properties
    ops = results.get("ops", None)
    if ops is not None:
        lines.append("# EFFECTIVE PROPERTIES (length = n)")
        for key in ("σ_s", "σ_m", "D_H2", "D_O2"):
            if key in ops:
                unit = {"σ_s":" [S/m]","σ_m":" [S/m]","D_H2":" [m^2/s]","D_O2":" [m^2/s]"}.get(key,"")
                lines.append(_summ_stats(ops[key], key, unit))
        lines.append("")
        lines.append("# MATRICES")
        for key in ("A_s", "A_m", "A_H2", "A_O2"):
            if key in ops and hasattr(ops[key], "shape"):
                A = ops[key]
                lines.append(f"{key}: shape={A.shape}, nnz={A.nnz}")
        lines.append("")

    # ---- Volume fractions (computed from params & mesh so they always exist)
    if (params is not None) and (mesh is not None):
        ε_cat, ε_ion, ε_pore, ε_H2O_ion = compute_volume_fractions(params, mesh)
        lines.append("# VOLUME FRACTIONS")
        lines.append(_summ_stats(ε_cat, "ε_cat"))
        lines.append(_summ_stats(ε_ion, "ε_ion"))
        lines.append(_summ_stats(ε_pore, "ε_pore"))
        lines.append(_summ_stats(ε_H2O_ion, "ε_H2O_ion"))
        lines.append("")

    # ---- Steady polarization results table again (nice alignment)
    if all(k in results for k in ("voltages","currents","converged","iterations")):
        vs = results["voltages"]
        cs = results["currents"]
        gs = results["converged"]
        it = results["iterations"]
        lines.append("# STEADY-STATE POLARIZATION (EASY TABLE)")
        lines.append(f"{'Voltage[V]':<12} | {'Current[A/cm^2]':<18} | {'Converged':<10} | {'Iters':<7}")
        lines.append("-"*60)
        for V, j, g, niter in zip(vs, cs, gs, it):
            jtxt = f"{j:.6g}" if (j is not None and np.isfinite(j)) else "nan"
            lines.append(f"{V:<12.4f} | {jtxt:<18} | {str(g):<10} | {int(niter):<7}")
        lines.append("")

    # ---- Transient summary
    tr = results.get("transient", None)
    if tr is not None:
        lines.append("# TRANSIENT SUMMARY")
        max_H2 = float(np.max(tr.H2_in_O2_channel)) if tr.H2_in_O2_channel.size else float("nan")
        steady_initial = float(tr.H2_in_O2_channel[0]) if tr.H2_in_O2_channel.size else float("nan")
        steady_final = float(tr.H2_in_O2_channel[-1]) if tr.H2_in_O2_channel.size else float("nan")
        lines.append(f"t span: [{tr.t[0]:.3f}, {tr.t[-1]:.3f}] s  (N={len(tr.t)})")
        lines.append(f"H2-in-O2: initial={steady_initial:.6g} vol%, peak={max_H2:.6g} vol%, final={steady_final:.6g} vol%")
        # Include a few time samples for H2-in-O2 signals
        sample_idx = np.linspace(0, len(tr.t)-1, num=min(10, len(tr.t)), dtype=int)
        lines.append("Sampled H2-in-O2 (channel) [t[s], value(vol%)]:")
        for k in sample_idx:
            lines.append(f"  {tr.t[k]:.3f}, {tr.H2_in_O2_channel[k]:.6g}")
        lines.append("")

    # ---- Plot path (if saved)
    if "plot_path" in results and results["plot_path"]:
        lines.append("# PLOT")
        lines.append(f"plot_path: {results['plot_path']}")
        lines.append("")

    # ---- Write file
    with open(filename, "w", encoding="utf-8") as f:
        f.write("\n".join(lines))

    # Small companion CSV for the steady curve (useful for quick plotting in other tools)
    try:
        if all(k in results for k in ("voltages","currents","converged","iterations")):
            vs = np.array(results["voltages"], dtype=float)
            cs = np.array(results["currents"], dtype=float)
            gs = np.array(results["converged"], dtype=bool)
            it = np.array(results["iterations"], dtype=int)
            csv_name = os.path.splitext(filename)[0] + "_polarization.csv"
            with open(csv_name, "w", encoding="utf-8") as f:
                f.write("Voltage_V,Current_A_per_cm2,Converged,Iters\n")
                for V, j, g, niter in zip(vs, cs, gs, it):
                    jtxt = "" if (j is None or not np.isfinite(j)) else f"{j:.12g}"
                    f.write(f"{V:.8f},{jtxt},{int(g)},{int(niter)}\n")
    except Exception:
        pass

    print(f"\nDebug report saved: {os.path.abspath(filename)}")
    return filename

# Automatically write the report after main() finishes
if __name__ == "__main__":
    try:
        _ = write_debug_report(results)  # default filename with timestamp
    except Exception as e:
        print(f"[debug report] Failed to write report: {e}")
```
